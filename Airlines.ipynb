## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
fightpred['airline'] = le.fit_transform(fightpred['airline'])
fightpred['source_city'] = le.fit_transform(fightpred['source_city'])
fightpred['stops'] = le.fit_transform(fightpred['stops'])
fightpred['departure_time'] = le.fit_transform(fightpred['departure_time'])
fightpred['destination_city'] = le.fit_transform(fightpred['destination_city'])
fightpred['arrival_time'] = le.fit_transform(fightpred['arrival_time'])
fightpred['class'] = le.fit_transform(fightpred['class'])

fightpred.info()
```

```python
fightpred.drop('flight',  axis = 1, inplace = True)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
x = fightpred.drop('price' , axis = 1)
```

```python
y = fightpred['price']
```

```python
y
```

```python
vif_data = pd. DataFrame()
vif_data['Columns']= x.columns
vif_data['VIF_values']= [variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y, train_size=0.8, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr= LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
difference= pd.DataFrame(np.c_[y_test,y_pred],columns= ["Actual Values", "Pred Values"])
```

```python
difference
```
